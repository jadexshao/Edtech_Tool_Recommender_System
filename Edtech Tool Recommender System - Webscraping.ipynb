```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import time
from IPython.core.display import clear_output

urls_list = pd.read_csv('Edtech Product URLs.csv')
urls = urls_list

list_cut =urls.transpose().values.tolist()

links = [val for sublist in list_cut for val in sublist]
```

```python



products = []

def get_product_name(soup):
    try:
        product_name = soup.h1.string
        products.append(product_name)
    except AttributeError:
        products.append('None')
    

```

```python

grades_list = []
def get_grade(soup):
    try:
        grade = soup.find('div', attrs={'class':'product-grades__value'})
        grades= (grade.find('span')).string
        grades_list.append(grades)
    
    except AttributeError: 
        grades_list.append('None')

```

```python

prices = []
def get_prices(soup):
    try:
        price = soup.find('span', attrs = {'class': 'product-price__value'}).string
        prices.append(price)
        
    except AttributeError: 
        prices.append('None')
```

```python

skills = []
subjects_list = []
def get_subjects_skills(soup):
    try: 
        skill = soup.find('div',attrs = {'class': 'subjects-skills__item'})
        skill_ = (skill.find('span')).string
        skills.append(skill_)
    
    except AttributeError: 
        skills.append('None')
    
    try:
        subject = skill.find_next('div',attrs = {'class': 'subjects-skills__item'})
        subjects = (subject.find('span')).string
        subjects_list.append(subjects)
        
    except AttributeError:
        subjects_list.append('N/A')
```

```python
p_descriptions = []
def get_descriptions(soup):
    try:
        p_des = soup.find('div',attrs = {'id': 'collapse-tool-overview'})
        product_description = (p_des.find('p')).text
        p_descriptions.append(product_description)
        
    except AttributeError:
        p_descriptions.append('N/A')
```

```python
teachers = []
def get_teachers(soup):
    try: 
        teacher = soup.find('a',attrs = {'class':'mobile-break teacher-name'}).string
        teachers.append(teacher)
    except AttributeError: 
        teachers.append('None')

```

```python
schools = []
def get_schools(soup):
    try:
        school = soup.find('div', attrs = {'class': 'school-info mobile-hide'}).text
        s_sp = school.split()
        school_name = s_sp[0]+" "+ s_sp[1] + " " + s_sp[2]
        schools.append(school_name)
    
    except (AttributeError, IndexError): 
        schools.append('N/A')

        
        
```

```python
ratings =[]
reviews = []
def get_ratings_reviews(soup):
    try: 
        rating = soup.find('div', attrs = {'class': 'teacher-review-title'})
        teacher_rating = rating.find_previous('span').string
        ratings.append(teacher_rating)
    except AttributeError:
        ratings.append('N/A')
        
    try:
        review = rating.find_next('p')
        teacher_review = review.text
        reviews.append(teacher_review)
    except AttributeError:
        reviews.append('N/A')
 
```

```python
sucesses = 0

for url in links: 
    page = requests.get(url)
    
    time.sleep(10)
    sucesses += 1

    print('Request:{}'.format(sucesses))
    clear_output(wait = True)
    
    soup = BeautifulSoup(page.text,'html.parser')
    get_product_name(soup)
    get_grade(soup)
    get_prices(soup)
    get_teachers(soup)
    get_subjects_skills(soup)
    get_descriptions(soup)
    get_schools(soup)
    get_ratings_reviews(soup)
    




        
```

```python
import pandas as pd


edtech_data = pd.DataFrame(products)
edtech_data.columns = ['product_name']
edtech_data['grades'] = grades_list
edtech_data['price'] = prices
edtech_data['skills'] = skills
edtech_data['subjects'] = subjects_list
edtech_data['product description'] = p_descriptions
edtech_data['teacher'] = teachers
edtech_data['school'] = schools
edtech_data['teacher rating'] = ratings
edtech_data['teacher review'] = reviews

edtech_data



```

```python

edtech_data.to_csv('Edtech Recommender Webscraping Data.csv', mode='a', header=False, index = False)
```
